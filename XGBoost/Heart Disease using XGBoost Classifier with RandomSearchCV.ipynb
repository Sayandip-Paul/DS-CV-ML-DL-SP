**Dataset Decsription**

The heart disease dataset you provided appears to contain information about individuals and various attributes related to heart disease. Let's go through each column and understand what the values represent:

1. age: The age of the individual (in years).
2. sex: The sex of the individual (0 = female, 1 = male).
3. cp: Chest pain type experienced by the individual. This is categorized into four types (0, 1, 2, 3).
4. trestbps: Resting blood pressure of the individual (in mm Hg).
5. chol: Serum cholesterol level of the individual (in mg/dl).
6. fbs: Fasting blood sugar level of the individual. This indicates whether the fasting blood sugar is greater than 120 mg/dl (1 = true, 0 = false).
7. restecg: Resting electrocardiographic results. This is categorized into three values (0, 1, 2).
8. thalach: Maximum heart rate achieved by the individual during exercise.
9. exang: Exercise-induced angina (1 = yes, 0 = no).
10. oldpeak: ST depression induced by exercise relative to rest.
11. slope: The slope of the peak exercise ST segment. This is categorized into three values (0, 1, 2).
12. ca: The number of major vessels (0-3) colored by fluoroscopy.
13. thal: Thalassemia, a blood disorder, categorized into three values (0, 1, 2, 3).
14. target: The presence of heart disease (0 = no heart disease, 1 = heart disease).

Each row in the dataset represents a different individual, and the values in each column provide information about their characteristics and test results.

```python
X = df.drop('target',axis = 1)
y = df['target']
```

```python
df.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
params = {
    "learning_rate" : [0.05,0.10,0.15,0.20,0.25,0.30],
    "max_depth" : [3,4,5,6,8,10,12,15],
    "min_child_weight" : [1,3,5,7],
    "gamma" : [0.0,0.1,0.2,0.3,0.4],
    "colsample_bytree" : [0.3,0.4,0.5,0.7]
}
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
model = xgb.XGBClassifier()
```

```python
tuned_model = RandomizedSearchCV(model, param_distributions = params, n_iter = 5, scoring = 'roc_auc', n_jobs = -1, cv = 10, verbose = 3)
tuned_model.fit(X_train,y_train)
```

```python
tuned_model.best_estimator_
```

```python
tuned_model.best_params_
```

```python
tuned_hyper_model = xgb.XGBClassifier(min_child_weight = 1,
 max_depth = 10,
 learning_rate = 0.25,
 gamma = 0.0,
 colsample_bytree = 0.3)
```

```python
tuned_hyper_model.fit(X_train,y_train)
y_pred = tuned_hyper_model.predict(X_test)
```

```python
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
